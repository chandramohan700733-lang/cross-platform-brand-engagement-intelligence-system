```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('cleaned_social_media_engagement.csv')
platform_engagement = df1.groupby('platform')['total_engagement'].mean().sort_values(ascending=False)
print(platform_engagement)
```

```python
import matplotlib.pyplot as plt

platform_engagement = df1.groupby('platform')['total_engagement'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
platform_engagement.plot(kind='bar')
plt.title('Average Total Engagement by Platform')
plt.xlabel('Platform')
plt.ylabel('Average Total Engagement')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
topic_engagement = df1.groupby('topic_category')['total_engagement'].mean().sort_values(ascending=False)
print(topic_engagement)
```

```python
plt.figure(figsize=(10,5))
topic_engagement.plot(kind='bar')
plt.title('Average Total Engagement by Topic Category')
plt.xlabel('Topic Category')
plt.ylabel('Average Total Engagement')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sentiment_engagement = df1.groupby('sentiment_bucket')['total_engagement'].mean().reindex([
    'Strong Negative', 'Negative', 'Neutral', 'Positive', 'Strong Positive'
])
print(sentiment_engagement)
```

```python
plt.figure(figsize=(8,5))
sentiment_engagement.plot(kind='bar')
plt.title('Average Total Engagement by Sentiment Bucket')
plt.xlabel('Sentiment Bucket')
plt.ylabel('Average Total Engagement')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
campaign_phase_perf = df1.groupby('campaign_phase')['engagement_rate'].mean().sort_values(ascending=False)
print(campaign_phase_perf)
```

```python
plt.figure(figsize=(8,5))
campaign_phase_perf.plot(kind='bar')
plt.title('Average Engagement Rate by Campaign Phase')
plt.xlabel('Campaign Phase')
plt.ylabel('Average Engagement Rate')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
