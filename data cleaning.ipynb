```python
def get_sentiment_bucket(score):
    if score <= -0.6:
        return 'Strong Negative'
    elif score <= -0.2:
        return 'Negative'
    elif score < 0.2:
        return 'Neutral'
    elif score < 0.6:
        return 'Positive'
    else:
        return 'Strong Positive'

df['sentiment_bucket'] = df['sentiment_score'].apply(get_sentiment_bucket)

def get_toxicity_flag(score):
    if score >= 0.75:
        return 'High Risk'
    elif score >= 0.4:
        return 'Medium Risk'
    else:
        return 'Low Risk'

df['toxicity_flag'] = df['toxicity_score'].apply(get_toxicity_flag)

df['engagement_per_1000_impressions'] = (df['total_engagement'] / df['impressions']) * 1000

print(df[['sentiment_score', 'sentiment_bucket', 'toxicity_score', 'toxicity_flag', 'engagement_per_1000_impressions']].head())
```

```python
print(df['sentiment_bucket'].value_counts())
print(df['toxicity_flag'].value_counts())
print(df['engagement_per_1000_impressions'].describe())
```

```python
df['virality_score'] = (
    df['shares_count'] * 0.5 + 
    df['comments_count'] * 0.3 + 
    df['likes_count'] * 0.2
)
engagement_median = df['engagement_per_1000_impressions'].median()
print(engagement_median)
def get_engagement_quality(score):
    if score >= engagement_median:
        return 'High Quality Engagement'
    else:
        return 'Low Quality Engagement'

df['engagement_quality_flag'] = df['engagement_per_1000_impressions'].apply(get_engagement_quality)
print(df[['likes_count', 'shares_count', 'comments_count', 'virality_score', 'engagement_quality_flag']].head())
print(df['engagement_quality_flag'].value_counts())
print(df['virality_score'].describe())
```
